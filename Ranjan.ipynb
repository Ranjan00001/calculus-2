```python
import math
A = 1
B = 2
C = 1
x_0 = 1
v_0 = 1
```

```python
D = B**2-4*A*C
# print(D)
alpha = B/(2*A)
beta = (abs((B**2)/(4*A**2)-(C/A)))**0.5
a1 = (-B + D**0.5)/2*A
a2 = (-B - D**0.5)/2*A 
# print(alpha, beta, a1, a2)    #a1,a2 are roots of the quadratic equation formed
```

```python
tlist = [i/100 for i in range(1001)]
xlist = []
# print(v_0,x_0,a1,a2)
# print(D,alpha,beta)

if D > 0:
  print('over-damping')
  d1 = (v_0 - x_0*a1)/(a1 - a2)
  d2 = (v_0 - x_0*a2)/(a2 - a1)
  # print(d1,d2)
  for t in tlist:
    x = (d1*math.exp((-alpha + beta)*t)+d2*math.exp((-alpha-beta)*t))
    xlist.append(x)
elif D < 0:
  print('under-damping')
  for t in tlist:
    x = math.exp(-alpha*t)*(x_0*math.cos(beta*t)+((v_0+alpha*x_0)*math.sin(beta*t)/beta))
    xlist.append(x)
else:
  print('critically-damping')
  for t in tlist:
    x = math.exp(-B*t/(2*A))*(x_0 + (v_0-a1*x_0)*t)
    xlist.append(x)
# print(xlist)
```

```python
import plotly.express as px
fig = px.line(x = tlist, y = xlist)
fig.show()

```
